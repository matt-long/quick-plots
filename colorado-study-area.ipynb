```python
%matplotlib inline

import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.io.shapereader as shpreader
```

```python
fig = plt.figure()

ax = plt.axes(projection=ccrs.NearsidePerspective(
                        central_latitude=30,
                        central_longitude=-100,
                        satellite_height=10000000.0))
ax.coastlines(resolution='110m')
ax.gridlines()

shapename = 'admin_1_states_provinces_lakes_shp'
states_shp = shpreader.natural_earth(resolution='110m',
                                     category='cultural', name=shapename)


for astate in shpreader.Reader(states_shp).records():

    facecolor = 'w'    
    if astate.attributes['name'] == 'Colorado':
        facecolor = 'r'          
        
    ax.add_geometries([astate.geometry], ccrs.PlateCarree(),
                      facecolor=facecolor, edgecolor='k')

plt.savefig('colorado-study-area.pdf', dpi=300, bbox_inches='tight')
```
