```python
%matplotlib inline
import os
import matplotlib.pyplot as plt
import cartopy.crs as ccrs


os.environ['CARTOPY_USER_BACKGROUNDS'] = './data'
```

```python
fig = plt.figure(figsize=(10, 10))

ax = plt.axes(projection=ccrs.Orthographic(
                        central_longitude=-90,
                        central_latitude=10,
                        )
             )


ax.set_global()
ax.coastlines(resolution='110m', linewidth=0.6)
ax.background_img('natural-earth-1', resolution='small')


# define list of lon, lat points
points = [
    (-126.72, 50.95),
    (-78.89, 0.165),
    (-72.56, -51.06),
]
for x, y in points:
    ax.plot(x, y, '*',
            color='b',
            markersize=12,
            transform=ccrs.PlateCarree(),
           )

plt.savefig('american-cordillera-sites.pdf', dpi=300, bbox_inches='tight')
```
